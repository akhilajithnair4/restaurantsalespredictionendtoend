# Exploring the Data

Check Various Categories Present in the Different Catrgorical Column

```python
for column in list(data.columns):
    if data[column].dtype==object and len(data[column].unique())<=10:
        
        print("categories in {0} variable are {1}".format(column,data[column].unique()))
```

Define the numerical and Categorical Variables

```python
data.drop(columns="Name",inplace=True)
```

```python
num_columns=[num_column for num_column in list(data.columns) if data[num_column].dtype!=object]
```

```python
cat_columns=[cat_column for cat_column in list(data.columns) if data[cat_column].dtype==object]
```

```python
num_columns
```

```python
cat_columns
```

```python
print("We have {} numerical columns,{}".format(len(num_columns),num_columns))
```

```python
print("We have {} categorical columns,{}".format(len(cat_columns),cat_columns))
```

```python
corr=data.corr()
```

```python
sns.heatmap(corr,annot=True, fmt=".1f")
plt.show()
```

```python
plt.figure(figsize=(20,8))
corr['Revenue'].sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(12, 8))
ax = sns.boxplot(data=data, palette='Pastel1')

# Customize the plot
ax.set_title('Customized Boxplot', fontsize=20, weight='bold')
ax.set_xlabel('Categories', fontsize=15)
ax.set_ylabel('Values', fontsize=15)

# Rotate x-axis labels
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)

# Customize grid lines
ax.grid(True, linestyle='--', alpha=0.7)

# Show plot
plt.tight_layout()
plt.show()
```
